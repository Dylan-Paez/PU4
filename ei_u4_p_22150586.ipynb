# Introduccion

El presente estudio analiza el impacto de diversas variables sociales, familiares y personales sobre el desempeño académico de estudiantes, medido a través de los puntajes en exámenes finales. Para ello, se emplearon herramientas estadísticas como ANOVA y la prueba de Tukey, con el fin de identificar diferencias significativas entre grupos categorizados según factores como participación parental, acceso a recursos, actividades extracurriculares, nivel de motivación, entre otros. Este enfoque busca ofrecer una comprensión más profunda de los elementos que influyen en los resultados académicos, sirviendo como base para formular recomendaciones que fortalezcan las políticas educativas y promuevan la igualdad de oportunidades.

El análisis incluye 13 variables, todas consideradas relevantes para evaluar la complejidad de los procesos educativos. Entre ellas, destacan el nivel socioeconómico, la calidad del profesorado, el tipo de escuela y las dinámicas de interacción social. Este enfoque integral permite una evaluación precisa de los factores que afectan el rendimiento, proporcionando insights que pueden guiar a los responsables de la toma de decisiones en educación.

# Marco Teorico

El rendimiento académico es un indicador clave en la evaluación del éxito educativo y se ve influenciado por una combinación de factores personales, familiares y escolares. Según teorías educativas como las de Bronfenbrenner (1979), el entorno social inmediato y las políticas públicas tienen un impacto directo en el aprendizaje. La participación parental, por ejemplo, fomenta un ambiente de apoyo que refuerza la motivación intrínseca y extrínseca del estudiante. Adicionalmente, el acceso a recursos educativos y la calidad del profesorado desempeñan roles fundamentales en el desarrollo de habilidades y conocimientos.

El uso de análisis estadísticos como ANOVA y Tukey permite desentrañar relaciones complejas entre estas variables. ANOVA evalúa las diferencias significativas entre grupos, mientras que Tukey identifica los pares específicos en los que se presentan discrepancias. Estas herramientas garantizan un enfoque objetivo y basado en evidencia para interpretar los datos y proponer mejoras educativas.



# Metodologia

Se realizó un análisis basado en datos de estudiantes que incluían 13 variables relacionadas con factores sociales, familiares y personales. Se emplearon las siguientes técnicas:

ANOVA: Para determinar si existían diferencias significativas en los puntajes de exámenes entre niveles de cada variable.

Tukey: Para evaluar cuáles pares de grupos presentaban diferencias significativas.

```python
#Dylan Sebastian Paez Serrano
import pandas as pd

# Cargar el dataset
df = pd.read_csv("https://raw.githubusercontent.com/Dylan-Paez/Proyecto-unidad-4/refs/heads/main/dataset.csv")

# Especificar las columnas que deseas conservar
columnas_deseadas = [
    'participacion_parental','acceso_a_recursos','actividades_extracurriculares','nivel_motivacion','acceso_internet','ingreso_familiar',
    'calidad_profesorado','tipo_escuela', 'influencia_companieros', 'discapacidad_aprendizaje','nivel_estudio_padres',
    'distancia', 'genero', 'puntaje_examen'
]

# Seleccionar solo las columnas deseadas
df_filtrado = df[columnas_deseadas]

# Eliminar filas con valores nulos
df_filtrado.dropna(inplace=True)

# Mostrar el DataFrame resultante
df=df_filtrado
df
```

# 1.Participacion_parental

```python
df['participacion_parental'].unique()
```

# **Anova**

```python
#Dylan Sebastian Paez Serrano
# Si todos los grupos tienen una distribución normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado'] ['puntaje_examen'],
)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadístico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusión: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusión: No hay diferencias significativas entre los tratamientos.")
```

**Interpretación del ANOVA:**

El análisis ANOVA mostró un F-Estadístico de 80.46 y un valor p de 0.00000, lo que indica que existen diferencias significativas entre los tres niveles de participación parental (bajo, medio y elevado).

# **Tukey**

```python
#Dylan Sebastian Paez Serrano
# Si todos los grupos tienen una distribución normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los gráficos de Tukey para tomar una decisión (ignorando la tabla).
#.    Estrictamente hablando, tendrían que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

**Interpretación de Tukey:**

Tukey confirmó que los estudiantes con participación parental "elevada" obtuvieron puntajes significativamente más altos que aquellos con niveles "medio" y "bajo". Además, los estudiantes con participación "media" también superaron a los de nivel "bajo".
Conclusión:

# **Conclusión**



Los resultados reflejan que un mayor nivel de participación parental contribuye significativamente a un mejor rendimiento académico. Esto sugiere que el apoyo de los padres es crucial para el éxito escolar.

```python
#Dylan Sebastian Paez Serrano
# Si todos los grupos tienen una distribución normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los gráficos de Tukey para tomar una decisión (ignorando la tabla).
#.    Estrictamente hablando, tendrían que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['discapacidad_aprendizaje'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```
