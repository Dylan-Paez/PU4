# Introduccion

El presente estudio analiza el impacto de diversas variables sociales, familiares y personales sobre el desempeo acadmico de estudiantes, medido a travs de los puntajes en exmenes finales. Para ello, se emplearon herramientas estadsticas como ANOVA y la prueba de Tukey, con el fin de identificar diferencias significativas entre grupos categorizados segn factores como participacin parental, acceso a recursos, actividades extracurriculares, nivel de motivacin, entre otros. Este enfoque busca ofrecer una comprensin ms profunda de los elementos que influyen en los resultados acadmicos, sirviendo como base para formular recomendaciones que fortalezcan las polticas educativas y promuevan la igualdad de oportunidades.

El anlisis incluye 13 variables, todas consideradas relevantes para evaluar la complejidad de los procesos educativos. Entre ellas, destacan el nivel socioeconmico, la calidad del profesorado, el tipo de escuela y las dinmicas de interaccin social. Este enfoque integral permite una evaluacin precisa de los factores que afectan el rendimiento, proporcionando insights que pueden guiar a los responsables de la toma de decisiones en educacin.

# Marco Teorico

El rendimiento acadmico es un indicador clave en la evaluacin del xito educativo y se ve influenciado por una combinacin de factores personales, familiares y escolares. Segn teoras educativas como las de Bronfenbrenner (1979), el entorno social inmediato y las polticas pblicas tienen un impacto directo en el aprendizaje. La participacin parental, por ejemplo, fomenta un ambiente de apoyo que refuerza la motivacin intrnseca y extrnseca del estudiante. Adicionalmente, el acceso a recursos educativos y la calidad del profesorado desempean roles fundamentales en el desarrollo de habilidades y conocimientos.

El uso de anlisis estadsticos como ANOVA y Tukey permite desentraar relaciones complejas entre estas variables. ANOVA evala las diferencias significativas entre grupos, mientras que Tukey identifica los pares especficos en los que se presentan discrepancias. Estas herramientas garantizan un enfoque objetivo y basado en evidencia para interpretar los datos y proponer mejoras educativas.

```python
#Dylan Sebastian Paez Serrano
import pandas as pd

# Cargar el dataset
df = pd.read_csv("https://raw.githubusercontent.com/Dylan-Paez/Proyecto-unidad-4/refs/heads/main/dataset.csv")

# Especificar las columnas que deseas conservar
columnas_deseadas = [
    'participacion_parental','acceso_a_recursos','actividades_extracurriculares','nivel_motivacion','acceso_internet','ingreso_familiar',
    'calidad_profesorado','tipo_escuela', 'influencia_companieros', 'discapacidad_aprendizaje','nivel_estudio_padres',
    'distancia', 'genero', 'puntaje_examen'
]

# Seleccionar solo las columnas deseadas
df_filtrado = df[columnas_deseadas]

# Eliminar filas con valores nulos
df_filtrado.dropna(inplace=True)

# Mostrar el DataFrame resultante
df=df_filtrado
df
```

# 1.Participacion_parental

```python
df['participacion_parental'].unique()
```

# **Anova**

```python
#Dylan Sebastian Paez Serrano
# Si todos los grupos tienen una distribucin normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado'] ['puntaje_examen'],
)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadstico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusin: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusin: No hay diferencias significativas entre los tratamientos.")
```

**Interpretacin del ANOVA:**

El anlisis ANOVA mostr un F-Estadstico de 80.46 y un valor p de 0.00000, lo que indica que existen diferencias significativas entre los tres niveles de participacin parental (bajo, medio y elevado).

# **Tukey**

```python
#Dylan Sebastian Paez Serrano
# Si todos los grupos tienen una distribucin normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los grficos de Tukey para tomar una decisin (ignorando la tabla).
#.    Estrictamente hablando, tendran que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Grfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Mltiples comparaciones entre todos los pares (Tukey)")
```

**Interpretacin de Tukey:**

Tukey confirm que los estudiantes con participacin parental "elevada" obtuvieron puntajes significativamente ms altos que aquellos con niveles "medio" y "bajo". Adems, los estudiantes con participacin "media" tambin superaron a los de nivel "bajo".
Conclusin:

# **Conclusin**



Los resultados reflejan que un mayor nivel de participacin parental contribuye significativamente a un mejor rendimiento acadmico. Esto sugiere que el apoyo de los padres es crucial para el xito escolar.

# **Anova**

```python
#Dylan Sebastian Paez Serrano
# Si todos los grupos tienen una distribucin normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['discapacidad_aprendizaje'] == 'no']['puntaje_examen'],
    df[df['discapacidad_aprendizaje'] == 'si']['puntaje_examen'],
)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadstico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusin: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusin: No hay diferencias significativas entre los tratamientos.")
```

**Interpretacin del ANOVA:**

Con un F-Estadstico de 45.21 y un valor p de 0.00000, ANOVA seal diferencias significativas entre estudiantes con y sin discapacidad de aprendizaje.

# **Tukey**

```python
#Dylan Sebastian Paez Serrano
# Si todos los grupos tienen una distribucin normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los grficos de Tukey para tomar una decisin (ignorando la tabla).
#.    Estrictamente hablando, tendran que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['discapacidad_aprendizaje'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Grfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Mltiples comparaciones entre todos los pares (Tukey)")
```

**Interpretacin de Tukey:**

La prueba de Tukey mostr que los estudiantes sin discapacidad de aprendizaje obtuvieron puntajes significativamente ms altos..

# **Conclusion**

Los estudiantes con discapacidades de aprendizaje enfrentan barreras adicionales que afectan su rendimiento, subrayando la necesidad de apoyo personalizado.
